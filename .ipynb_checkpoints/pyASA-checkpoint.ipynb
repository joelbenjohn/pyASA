### Download Results into Excel file

```python
!pip install xlsxwriter
import xlsxwriter
import Data
data1 = Data.DATA
#Name Document
workbook = xlsxwriter.Workbook('Truss.xlsx')
Nsheet = workbook.add_worksheet()
Esheet = workbook.add_worksheet()
Node_disp = workbook.add_worksheet()
Elem_stress = workbook.add_worksheet()
row = 0
# Print Node coordinate data
for col, data in enumerate(data1['xn']):
    Nsheet.write_column(row, col, data)
# Print Element Connectivity data
for col, data in enumerate(data1['ien']):
    Esheet.write_column(row, col, data)
# Print Node displacement data
for col, data in enumerate(data1['dcomp']):
    Node_disp.write_column(row, col, data)
# Print Element stresses data
st = data1['stress'].reshape(len(data1['stress']), 1)
for col, data in enumerate(st.T):
    Elem_stress.write_column(row, col, data)

workbook.close()
```
